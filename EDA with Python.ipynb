```python
#importing libraries
#Reading text file
import pandas as pd
import numpy as np
data = ['symboling','normolized_losses','make','fuel_type','aspiration','num_of_doors','body_style','drive_wheels','engine_location','wheel_base','length','width','height','curb_weight','engine_type','num_of_cylinders','engine_size','fuel_system','bore','stroke','compression_ratio','horsepower','peak_rpm','city_mpg','highway_mpg','price']
a = pd.read_csv(your_local_path+'Automobiledata1.txt')
a
```

```python
#cleaing the data of values = ? and changed datatype
a_n_l =a['normalized-losses'].loc[a['normalized-losses']!='?']
a_n_lmean= a_n_l.astype(str).astype(int).mean()
a['normalized-losses'] = a['normalized-losses'].replace('?',a_n_lmean).astype(int) 
a['normalized-losses'].head() 
a['horsepower'].str.isnumeric().value_counts()
a_hp =a['horsepower'].loc[a['horsepower']!='?']
a_hpmean=a_hp.astype(str).astype(int).mean()
a['horsepower'] = a['horsepower'].replace('?',a_hpmean).astype(int)
a['horsepower'].head()
a['price'].str.isnumeric().value_counts() 
a['price'].loc[a['price'].str.isnumeric()==False]
a_p=a['price'].loc[a['price']!='?']
a_pmean =a_p.astype(str).astype(int).mean()
a['price'] = a['price'].replace('?',a_pmean).astype(int)
a['price'].head()
a.dtypes
```

```python
#cleaing the data of values = ? and changed datatype
a[a['bore']!='?']
a['bore']=pd.to_numeric(a['bore'],errors='coerce')
a[a['stroke']!='?']
a['stroke']=pd.to_numeric(a['stroke'],errors='coerce')
a[a['peak-rpm']!='?']
a['peak-rpm']=pd.to_numeric(a['peak-rpm'],errors='coerce') 
a.dtypes
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

Number of Vehicles by Make using Bar graph 

```python
a.make.value_counts().nlargest(10).plot(kind='bar',figsize=(15,5))
plt.title("Number of Vehicles by Make")
plt.ylabel('Number of Vehicles')
plt.xlabel('make');
```

Number of vehicles by Fuel type

```python
a['fuel-type'].value_counts().plot(kind='bar',color ='blue',figsize=(10,7))
plt.title("Number of Vehicles by fuel type")
plt.ylabel('Number of Vehicles')
plt.xlabel('fuel type');
```

Number of Vehicles by body style

```python
a['body-style'].value_counts().plot(kind='bar',color ='orange',figsize=(10,10))
plt.title("Number of Vehicles by body style")
plt.ylabel('Number of Vehicles')
plt.xlabel('body style');
```

Number of vehicles by drive wheels

Bar diagram of city mpg Vs Drive wheels

Bar diagram of highway mpg Vs drive wheels
